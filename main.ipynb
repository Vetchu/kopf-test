```python
with open("fullfillment.json", "r", encoding="utf-8") as file:
    file_string = file.read()
b = json.loads(file_string)
b
```

```python
b["checkoutForms"][0]['lineItems'][0]
```

```python
order = b["checkoutForms"][0]
line_item = order['lineItems'][0]
line_item_name = line_item["offer"]["name"]
line_item_name
```

```python
names = []
for order in b["checkoutForms"]:
    for line_item in order["lineItems"]:
        offer = line_item["offer"]
        name = offer["name"]
        price = line_item["price"]["amount"]
        names.append({"name": name, "price": price})

names
```
